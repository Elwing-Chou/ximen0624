```python
from tensorflow.keras.datasets import mnist
# ((訓練圖片, 答案), (側視圖片, 答案))
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train[0]
```

```python
# 灰階圖片 0(黑) - 255(白)
import pandas as pd
pd.DataFrame(x_train[0])
```

```python
# model有兩種方式 Sequential(堆疊)  Model(複雜)
from tensorflow.keras.models import Sequential
# Dense: 全連接層(fully-connected)
from tensorflow.keras.layers import Dense

layers = [
    # 784 * 64 + 64(bias)
    Dense(64, activation="sigmoid", input_dim=784),
    # 標準化(加起來=1)的sigmoid -> softmax
    # 係數 64 * 10 + 10(bias)
    Dense(10, activation="softmax")
]
model = Sequential(layers)
model.summary()
```

```python
model.compile(loss="mse", metrics=["accuracy"])
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.utils import to_categorical
y_train_onehot = to_categorical(y_train)
y_test_onehot = to_categorical(y_test)
print(y_train[0])
print(y_train_onehot[0])
```

```python
# 其實係數不是完全隨機, 他是在某個範圍作隨機
# 其實變相要求你的輸入應該要落在某個範圍, tensorflow希望你的輸入落在0~1或者-1~1
x_train_scaled = x_train.reshape(60000, 784) / 255.0
x_test_scaled = x_test.reshape(10000, 784) / 255.0
```

```python
# batch_size: 看了多少筆才更新一次模型參數
# epochs: 整份資料要看幾round
# !!! 究竟是幾round??? 我們不會無限訓練下去(當然loss會越來越小)->過擬和
# 跟機器學習依樣, 你要準備一份他沒看過的, 來檢視現在的擬和程度
# batch=100 vali=0.1  1個epoch做了幾次模型更新: 54000 / 100 = 540
# 擬要看validation loss來決定訓練多少個epochs
# verbose: 印log程度 0(quiet) 1(bar) 2(沒有bar)
model.fit(x_train_scaled,
     y_train_onehot,
     batch_size=100,
     epochs=30,
     validation_split=0.1,
     verbose=2)
```

```python
# Demo argmax(之前的idxmax)
# axis=0: 直行的最大值 axis=1: 橫列最大值的位置
import numpy as np
np.array([
    [1, 2],
    [3, 4],
    [5, 6]
]).argmax(axis=1)
```

```python
model.evaluate(x_test_scaled, y_test_onehot)
```
